# **Merge Sort**

    [16,21,11,8,12,22] -> Merge Sort

* Aşamaları:

      1) İlk önce diziyi ortadan ikiye böleceğiz. Ve sağlı sollu iki ana gruba ayıracağız. Bölme işlemi bittikten sonra sıralanıp birleştirilecek.
          --> 16,21,11      ve    8,12,22
  
      2) İlk grup sola ikinci grup sağa yazılır(sağ sol farketmez). İlk grup kendi içinde ikiye bölünür. İkinci grup da öyle.
          --> 16,21   11    ve    8,12     22
              
      3) Sayılar tek kalana kadar bölünür.
          --> 16   21   11  ve    8   12   22

      4) Artık tek kalanlar, kendi grupları içinde sıralanıp olabildiğince ikili halde gruplandırılır.
          --> 16,21 ile 11  ve    8,12 ile 22

      5) İki ana grup birleştirilir tekrar, yine kendi içinde sıralama yapılarak.
          --> 11,16,21      ve    8,12,22

      6) Bu iki grup birleştirilip sıralanır ve son halini alır.
          --> 8,11,12,16,21,22

## **Big O Gösterimi:** O(n log n)
